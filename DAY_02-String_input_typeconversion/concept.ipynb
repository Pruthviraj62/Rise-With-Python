## Input and output , String and type conversion

### input output in python

```python
name = input("Enter your name : ")
print(f"Name is {name}")
```

### string indexing 

```python
a = 'A'
print(ord(a))

a = 65
print(chr(a))

print(chr(66))

print(ord('C'))
```

```python
a = "Hello Mauli"
print(a[0])
print(a[0:5:1])
print(a[-1])
print(a[-1:-6:-1])
print(a[5])
print(a[6::1])
print(a[::])
print(a[0::3])
```

### type conversion

```python
a = 12# a = []
# a = bool(a)
# print(a)
# print(type(a))
a = str(a)
print(type(a))

b = 12.4
b = int(b)
print(type(b))

c = 12
c = float(c)
print(type(c))

d = 20 
d = bool(d)
print(type(d))
print(d)

e = 0
print(bool(e))

#implicit type coversion
a = 12/3
print(type(a)) #--> automatically converted into float

```
