# Variables

```python
# variables are the containers to store data or value 
# we dont need to mention datatype while writing variable in python 

# Rules :  You can not use numbers at variable start
#  You can not use spaces in variables.
# You should not use special characters in variables.

# Example : 
name = "Pruthviraj Patil"
age = 20 

print(f"Name is {name}")
print(f"age is {age}")
```

## Datatypes

```python
# datatypes are type of data we are storing in a variable 
# Three types :
# 1] Number --> it consist (int , float , complex)
# 2] String 
# 3] Boolean

# Example : 
#1] Number 

number = 10
radius = 9.3
result = 4.9j

print(type(number))
print(type(radius))
print(type(result))

#2] String

name = "Pruthviraj"
print(type(name))

#3] Boolean 

value = True
print(type(value))
```
